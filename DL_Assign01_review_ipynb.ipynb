## Step 3. Define tts(text)

**def tts(text):**

- This line defines a function called tts that takes a text argument. The purpose of this function is to convert the given text into an audio file.
▶ 이 줄은 텍스트 인수를 받는 tts라는 함수를 정의한다. 이 기능의 목적은 주어진 텍스트를 오디오 파일로 변환하는 것이다.

**gtts_object = gTTS(text,lang = "en", slow = True)**

- This line creates a new gTTS object with the given text, specifying the language as English and the speech speed as slow. ▶ 이 줄은 지정된 텍스트를 새로운 gTTS 개체를 만들어 언어를 영어로 지정하고 음성 속도를 느리게 지정한다. 

**gtts_object.save("myaudio.mp3")**

- This line saves the audio file generated by the gTTS object to the file "myaudio.mp3" on the local filesystem. ▶ 이 줄은 gTTS에 의해 생성된 오디오 파일을 로컬 파일 시스템 "myaudio.mp3"파일에 저장한다.

**return Audio("myaudio.mp3")**

- This line returns an Audio object representing the audio file that was just saved. The Audio object can be used to play the audio file or manipulate it in various ways. ▶ 이 줄은 방금 저장된 오디오 파일을 나타내는 오디오 개체를 반환한다. 오디오 개체는 오디오 파일을 재생하거나 다양한 방법으로 조작하는 데 사용할 수 있다.

```python
def tts(text):
  gtts_object = gTTS(text,lang = "en", slow = True)
  gtts_object.save("myaudio.mp3")
  return Audio("myaudio.mp3")
```

This will generate an audio file containing the speech corresponding to the text "I'm reviewing DL I learned today" in English language spoken at a slow speed. # Using **input( )**

- In Python, input() is a built-in function that is used to get input from the user via the console. The input() function displays a prompt on the console, waits for the user to enter some text, and then returns the text entered by the user as a string.
▶ 파이선에서, input()은 콘솔을 통해 사용자로부터 입력을 받는 데 사용되는 내장 함수이다. Input()함수는 콘솔에 프롬포트를 표시하고 사용자가 텍스트를 입력할 때까지 기다린 다음 사용자가 입력한 텍스트를 문자열로 반환한다.
- The syntax of the input() function is simple. It takes an optional string argument that is displayed as a prompt to the user, and then waits for the user to enter some text. ▶ Input()함수의 구문은 간단하다. 사용자에게 프롬포트로 표시되는 선택적 문자열 인수를 사용한 다음 사용부가 일부 텍스트를 입력할 때까지 기다린다.
